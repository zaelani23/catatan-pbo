```python
class Rekening:
    def tambah_saldo(self):
        print("Saldo anda saat ini Rp.",self.saldo)
        self.uang = int(input("Masukkan jumlah uang"))
        self.saldo = self.saldo + self.uang
        print("Saldo anda saat ini Rp.",self.saldo)
    def penarikan(self):
        print("Saldo anda saat ini Rp.",self.saldo)
        self.uang = int(input("Masukkan jumlah uang"))
        self.saldo = self.saldo - self.uang
        print("Saldo anda saat ini Rp.",self.saldo)
    def pilih(self):
        while True:
            print("1. Setor tunai")
            print("2. Tarik tunai")
            pilihan = int(input("Masukan pilihan: "))
            if pilihan==1:
                self.tambah_saldo()
            elif pilihan==2:
                self.penarikan()
            else:
                print("menu tidak ada")
            ulang = input("Apakah anda ingin mengulang? Ya/Tidak")
            if ulang!="Ya":
                break
        
class Nasabah(Rekening):
    def __init__(self,nama,alamat,telp,rek,saldo):
        self.nama = nama
        self.alamat = alamat
        self.telp = telp
        self.rek = rek
        self.saldo = saldo
    def bio(self):
        print("Nama",self.nama)
        print("Alamat",self.alamat)
        print("Telp.",self.telp)
        print("Rekening",self.rek)
        
saya = Nasabah("Mohamad Zaelani","Nganjuk","09090909","918179172917",1000000)
saya.bio()
saya.pilih()
```

```python
from datetime import datetime
from dateutil import relativedelta
sekarang = datetime.now()
tahun_now = sekarang.year
bulan_now = sekarang.month
hari_now = sekarang.day
jam_now = sekarang.hour
menit_now = sekarang.minute
detik_now = sekarang.second
tgl_now = datetime(tahun_now,bulan_now,hari_now,jam_now,menit_now,detik_now)

class Orang:
    def __init__(self,nm,tgl,bln,thn):
        self.nama = nm
        self.tgl_lahir = datetime(thn,bln,tgl)
        
    def perkenalkan_anda(self):
        print("Halo, saya",self.nama)
        
    def prediksi_umur(self):
        self.preumur = relativedelta.relativedelta(tgl_now,self.tgl_lahir)
            
    def umur_sekarang(self):
        self.umurskrng = relativedelta.relativedelta(tgl_now,self.tgl_lahir)
        self.tahun = self.umurskrng.years
        self.bulan = self.umurskrng.months
        self.hari = self.umurskrng.days
        self.jam = self.umurskrng.hours
        self.menit = self.umurskrng.minutes
        self.detik = self.umurskrng.seconds
            
    def hasil(self):
        self.perkenalkan_anda()
        self.prediksi_umur()
        self.umur_sekarang()
        print("Perkiraan umur:",self.preumur.years,"tahun")
        print("Umur sekarang:",self.tahun,"tahun",self.bulan,"bulan",self.hari,"hari",self.jam,"jam",self.menit,"menit",self.detik,"detik\n")

o1 = Orang("Mohamad Zaelani",23,3,2000)
o1.hasil()

o2 = Orang("Yusuf Sugiono",24,1,2000)
o2.hasil()
```

```python
class Lingkaran:
    def __init__(self,jari):
        self.jari = jari
    def hitung(self):
        luasling = 3.14*self.jari*self.jari
        kelilingling = 3.14*2*self.jari
        print("Luas lingkaran adalah",luasling,"cm2")
        print("Keliling lingkaran adalah",kelilingling,"cm")
        
class Tabung(Lingkaran):
    def __init__(self,jari,tinggi):
        super().__init__(jari)
        self.tinggi = tinggi
    def hitung(self):
        super().hitung()
        luasper = 2*3.14*self.jari*(self.tinggi+self.jari)
        print("Luas permukaan tabung adalah",luasper,"cm2\n")
        
tabungku = Tabung(10,12)
tabungku.hitung()
```

```python
def fpb(a,b):
        if b == 0:
            return a
        else:
            return fpb(b,a % b)
class Pecahan:
    def __init__(self,pembilang,penyebut):
        self.pembilang = pembilang
        self.penyebut = penyebut
    def cetak(self):
        print("Pecahanya adalah "+str(int(self.pembilang))+"/"+str(int(self.penyebut)))
        self.new_pembilang = self.pembilang/fpb(self.pembilang,self.penyebut)
        self.new_penyebut = self.penyebut/fpb(self.pembilang,self.penyebut)
        print("Bentuk sederhananya adalah "+str(int(self.new_pembilang))+"/"+str(int(self.new_penyebut)))

coba = Pecahan(3,7)
coba.cetak()
data1 = Pecahan(6,142)
data1.cetak()
data2 = Pecahan(206,1048)
data2.cetak()
```

```python
class Mahasiswa:
    def __init__(self,nama,npm):
        self.nama = nama
        self.nip = npm
    def namanip(self):
        print("Nama:",self.nama)
        print("NIP:",self.nip)
class Matkul:
    def cetakmatkul(self,kode,mk):        
        self.kodemk = kode
        self.matkul = mk
        print("Nama Mata kuliah:",self.matkul)
        print("Kode Mata kuliah:",self.kodemk)
class AmbilMatkul(Mahasiswa,Matkul):
    def __init__(self,nama,npm,kode,mk,nilai):
        self.nama = nama
        self.nim = npm
        self.kdmk = kode
        self.mk = mk
        self.nilai = nilai
    def cetak(self):
        Mahasiswa.__init__(self,self.nama,self.nim)
        Mahasiswa.namanip(self)
        super().cetakmatkul(self.kdmk,self.mk)
        print("Nilai:",self.nilai)
        
saya = AmbilMatkul("Mohamad Zaelani","180441100030","SI101","PBO","A")
saya.cetak()       
```

```python
class Pegawai:
    def __init__(self,nama,nip,gol):
        self.nama = nama
        self. nip = nip
        self.gol = gol
        self.gp = 0
        if self.gol == 1:
            self.gp = 2000000
        elif self.gol == 2:
            self.gp = 3000000
        elif self.gol == 3:
            self.gp = 4000000
        elif self.gol == 4:
            self.gp = 5000000
        print("Nama",self.nama)
        print("NIP",self.nip)
        print("Golongan",self.gol)
class Karyawan(Pegawai):
    def __init__(self,nama,nip,gol,masuk):
        super().__init__(nama,nip,gol)
        self.masuk = masuk
    def hitunggaji(self):
        self.gt = self.gp*0.1*self.masuk
        print("Gaji pokok adalah",self.gp)
        print("Gaji tambahan adalah",self.gt)
        print("Gaji total adalah",self.gp+self.gt)
        print()
class Dosen(Pegawai):
    def __init__(self,nama,nip,gol,sks):
        super().__init__(nama,nip,gol)
        self.sks = sks
    def hitunggaji(self):
        self.gt = self.gp*0.1*self.sks
        print("Gaji pokok adalah",self.gp)
        print("Gaji tambahan adalah",self.gt)
        print("Gaji total adalah",self.gp+self.gt)
        print()
dosen1 = Dosen("Mohamad Zaelani","1817729",1,20)
dosen1.hitunggaji()
karyawan1 = Karyawan("Ani","182782791",4,25)
karyawan1.hitunggaji()
```

```python
class Kategori:
    def __init__(self,kategori):
        self.kategori = kategori
    def tampilkan(self):
        print("Kategori:",self.kategori)
class Barang(Kategori):
    def __init__(self,namabarang,kategori,kode,harga):
        self.namabarang = namabarang
        self.kode = kode
        self.harga = harga
        self.kategori = kategori
    def tampil(self):
        super().__init__(self.kategori)
        super().tampilkan()
        print("Nama barang:",self.namabarang)
        print("Kode barang:",self.kode)
        print("Harga Rp."+str(self.harga))
class Pembeli:
    def __init__(self,nama):
        self.nama = nama
    def tampilnama(self):
        print("Nama pembeli:",self.nama)
class TransaksiPembelian(Barang,Pembeli):
    def __init__(self,nama,namabarang,kategori,kode,harga):
        Pembeli.__init__(self,nama)
        Barang.__init__(self,namabarang,kategori,kode,harga)
    def transaksi(self):
        Pembeli.tampilnama(self)
        Barang.tampil(self)
        
ob = TransaksiPembelian("Mohamad Zaelani","Flash Disk 32Gb","Penyimpanan Komputer","Cd65SwrAA",75000)
ob.transaksi()
```
